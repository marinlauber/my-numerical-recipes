```python
import numpy as np
import matplotlib.pyplot as plt
```

[Kovasznay Flow](http://www.karlin.mff.cuni.cz/~prusv/vyuka/2016-2017/zs/downloads/kovasznay.pdf)

```python
def Kovasznay(x, y, Re):

    lam = Re/2. - ((Re**2)/4. + 4.*np.pi**2)**(.5)

    u = - np.exp(lam*x)*np.cos(2*np.pi*y[:, np.newaxis])

    v = -lam/(2*np.pi)*np.exp(lam*x)*np.sin(2*np.pi*y[:, np.newaxis])
    
    psi = y[:, np.newaxis] - 1./(2*np.pi)*np.exp(lam*x)*np.sin(2*np.pi*x*y[:, np.newaxis])
    
    omega = lam*Re*np.exp(lam*x)*np.sin(2*np.pi*y[:, np.newaxis])

    return u, v, psi, omega
```

```python
N = 1000
x = np.linspace(0, 1, N)
y = np.linspace(-1, 1., N)
u, v, psi, omega = Kovasznay(x, y, 0.1)
plt.figure(figsize=(8,8))
plt.streamplot(x, y, u, v, density=1.8);
```

```python
plt.contourf(x, y, omega, levels=31)
plt.colorbar();
```
